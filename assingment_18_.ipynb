```python
df['Time']=pd.to_datetime(df['Time'],format='%m/%d/%Y %H:%M')

df['month']=df['Time'].dt.month
df
```

```python
def is_dec(row):
  if row['month']==12:
    return 1
  else:
    return 0

df['IS_DECEMBER']=df.apply(is_dec,axis=1)
```

```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv('/content/visitor-interests.csv')
df
```

```python
pip install pyyaml ua-parser user-agents
```

```python
import user_agents
```

```python
ua=df['UserAgent']
```

```python
def get_browsers (br):
  ua= user_agents.parse(br)
  return ua.browser.family

def get_os (os_str):
  ua=user_agents.parse(os_str)
  return ua.os.family


def get_device (dvc):
  ua= user_agents.parse(dvc)
  return ua.device.family

df['Browser']=df['UserAgent'].apply(get_browsers)
df['OS']=df['UserAgent'].apply(get_os)
df['Device']=df['UserAgent'].apply(get_device) 
```

```python
df
```

```python
pip install ip2geotools
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
```

```python
def get_geo_data(ip):
  try:
    response=ip2geo.get(ip[:10],api_key='free')
    return pd.Series([response.latitude,response.longitude])
  except:
    return np.nan

df['latitude']=df['IP'].apply(get_geo_data)
df['longitude']=df['IP'].apply(get_geo_data)
```

1) Using visitor-interests.csv dataset, from the first 50 row extract latitude, longitude from 'IP' column and use it to create new 'Dist_From_Egypt' column calculating distance with Egypt Lat/Lonng, you can get this location from google.


```python
import pandas as pd 
import numpy as np
from geopy.distance import great_circle
```

```python
df.columns
```

```python
def get_dist(x):
    egypt_lat = 30.033333
    egypt_long = 31.233334
    egypt_loc = (egypt_lat, egypt_long)
    row_loc = (x['latitude'], x['longitude'])
    return great_circle(row_loc, egypt_loc).kilometers

df['Dist_From_Egypt'] = df.apply(get_dist, axis=1)
df.head()
```
